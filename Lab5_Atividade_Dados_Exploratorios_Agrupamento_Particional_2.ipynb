9. Para determinar o número de agrupamentos nos dados, podemos aplicar o k-medoid com um número variável de agrupamentos, de 2 a 10, e calcular a Soma das Distâncias Absolutas (SDA) correspondentes, como mostra neste exemplo:

```python
import matplotlib.pyplot as plt
from sklearn_extra.cluster import KMedoids

# Calcula Soma das Distâncias Absolutas (SDA) para os diferentes números de clusters (2-10)
sda_scores = []
for k in range(2, 11):
    kmedoids = KMedoids(n_clusters=k, random_state=0).fit(X)
    sda_scores.append(kmedoids.inertia_)

# Plot os valores de SDA
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), sda_scores, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('Soma das Distâncias Absolutas (SDA)')
plt.title('Agrupamento K-medoides')
plt.show()
```
